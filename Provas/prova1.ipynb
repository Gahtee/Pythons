```python
#Calculadora foda para pessoas com preguiça de fazer inúmeros calculos e com diversas funções variadas
#enfâse nas funções de Data e de MediaA

from datetime import datetime, timedelta

print(' $$$$$$\            $$\                     $$\                 $$\                                    $$$$$$$$\              $$\           ')
print('$$  __$$\           $$ |                    $$ |                $$ |                                   $$  _____|             $$ |          ')
print('$$ /  \__| $$$$$$\  $$ | $$$$$$$\ $$\   $$\ $$ | $$$$$$\   $$$$$$$ | $$$$$$\   $$$$$$\  $$$$$$\        $$ |    $$$$$$\   $$$$$$$ | $$$$$$\  ')
print('$$ |       \____$$\ $$ |$$  _____|$$ |  $$ |$$ | \____$$\ $$  __$$ |$$  __$$\ $$  __$$\ \____$$\       $$$$$\ $$  __$$\ $$  __$$ | \____$$\ ')
print('$$ |       $$$$$$$ |$$ |$$ /      $$ |  $$ |$$ | $$$$$$$ |$$ /  $$ |$$ /  $$ |$$ |  \__|$$$$$$$ |      $$  __|$$ /  $$ |$$ /  $$ | $$$$$$$ |')
print('$$ |  $$\ $$  __$$ |$$ |$$ |      $$ |  $$ |$$ |$$  __$$ |$$ |  $$ |$$ |  $$ |$$ |     $$  __$$ |      $$ |   $$ |  $$ |$$ |  $$ |$$  __$$ |')
print('\$$$$$$  |\$$$$$$$ |$$ |\$$$$$$$\ \$$$$$$  |$$ |\$$$$$$$ |\$$$$$$$ |\$$$$$$  |$$ |     \$$$$$$$ |      $$ |   \$$$$$$  |\$$$$$$$ |\$$$$$$$ |')
print(' \______/  \_______|\__| \_______| \______/ \__| \_______| \_______| \______/ \__|      \_______|      \__|    \______/  \_______| \_______|\n')

print('Digita qual operação quer fazer')
op=input('Calculadora\nIMC\nVDT\nGasolina\nDias\nÁgua\nMédiaA\nRegrade3: ')

if op == 'Calculadora' or op == 'calc':
  op2=input('Qual operação quer fazer? ')
  n1=float(input('Qual o primeiro número? '))
  n2=float(input('Qual o segundo número? '))
  if op2 == 'adc' or op2 == '+':
    print('A soma de {0} com {1} é'.format(n1, n2), n1+n2)
  elif op2 == 'sub' or op2 == '-':
    print('A subtração de {0} com {1} é'.format(n1, n2), n1-n2)
  elif op2 == 'mult' or op2 == '*':
    print('A multiplicação entre {0} e {1} é'.format(n1, n2), n1*n2)
  elif op2 == 'div' or op2 == '/':
    print('A divisão de {0} com {1} é'.format(n1, n2), n1/n2)
  elif op2 == 'pot' or op2 == '^':
    print('{0} elevado a {1} é'.format(n1, n2), n1**n2)
  elif op2 == 'raiz':
    print('{0} a raiz {1} é'.format(n1, n2), n1**(1/n2))
  elif op2 == 'porcentagem' or op2 == '%':
    print('{0}% de {1} é {2}'.format(n1,n2,(n1/100)*n2))
  else:
    print('Operação Invalida')
elif op == 'IMC':
  altura=float(input('Qual sua altura em metros? '))
  peso=float(input('Qual seu peso em quilos? '))
  imc=peso/(altura**2)
  if imc < 18.5:
    print('Peso abaixo do normal')
  elif imc >= 18.5 and imc < 25:
    print('Peso normal')
  elif imc >= 25 and imc < 30:
    print('Pré obesidade / acima do peso')
  elif imc >= 30 and imc < 35:
    print('Obesidade grau um')
  elif imc >= 35 and imc < 40:
    print('Obesidade grau dois')
  else:
    print('Obesidade grau três / mórbida')
elif op == 'VDT':
  op2 = input('Qual operação você deseja fazer?\n velocidade, tempo, distancia: ')
  if op2 == 'velocidade':
    d=float(input('Digite a variação da distancia (metros): '))
    t=float(input('Digite a variação do tempo (segundos): '))
    km=input('Prefere em km/h ou m/s? (k ou m): ')
    if km == 'k':
      k=(d/t)*3.6
      print('Sua velocidade media foi de {0}km/h'.format(k))
    else:
      print('Sua velocidade média foi de {0}m/s'.format(d/t))
  elif op2 == 'tempo':
    vm=float(input('Digite a velocidade (m/s): '))
    d=float(input('Digite a distancia (metros)): '))
    print('O tempo necessário é de {0} segundos\nOu {1} minutos'.format(d/vm,(vm/d)/60))
  elif op == 'distancia':
    t=float(input('Digite a variação do tempo (segundos): '))
    vm=float(input('Digite a velocidade (m/s): '))
    print('A distancia percorrida foi de {0} metros ou {1} quilometros'.format(t*vm, (t*vm)/1000))
  else:
    print('Operação invalida')
elif op == 'Gasolina':
  c=float(input('Quantos km seu carro faz por litro? '))
  dp=float(input('Quantos quilometros foram percorridos? '))
  pg=float(input('Qual preço da gasolina? '))
  print('Você gastou {0} R$ nessa viagem'.format((dp/c)*pg))
elif op == 'Dias':
  dt1=input('Digite a primeira data (inicial): ')
  dt2=input('Digite a segunda data (final): ')
  tsdt1 = datetime.strptime(dt1, "%d/%m/%Y")
  tsdt2 = datetime.strptime(dt2, "%d/%m/%Y")
  difi= (tsdt2-tsdt1)
  diast= difi.days
  anos= diast//365
  meses= (diast-anos*365)//30
  dias= (diast-anos*365 - meses*30)
  print('Se passaram {0} anos, {1} meses, {2} dias'.format(anos, meses, dias))
elif op == 'Agua' or op == 'Água':
  peso=float(input('Qual seu peso? '))
  print('A quantidade de agua ideal é de {0} litros'.format((peso*40)/1000))
elif op == 'MediaA' or op == 'MédiaA':
  qnt=int(input('Quantos termos você quer a média? '))
  m=qnt
  x=0
  while (qnt != 0):
    n=float(input('Digite o {0:d}º numero: '.format((m-qnt)+1)))
    x=x+(n/m)
    qnt=qnt-1
  print('A média aritimetica foi de: ', x)
elif op == 'Regrade3' or op == 'regrade3':
  print('N1 --- N2\nX  --- N3')
  n1=float(input('Digite o N1: '))
  n2=float(input('Digite o N2: '))
  n3=float(input('Digite o N3: '))
  x=(n1*n3)/n2
  print('O resultado de X foi:', x)
else:
  print('Opção invalida')
```
