```python
import PIL.Image
import requests
import urllib.request
import json
from io import BytesIO
with urllib.request.urlopen('http://api.nekos.fun:8080/api/lewd')  as response:
  neko = json.load(response)
response = requests.get(neko['image'])
img = PIL.Image.open(BytesIO(response.content))
display(img)
```
